# weatherisk Quickstart (Python Pipeline)

This notebook runs a lightweight end-to-end synthetic validation of the Python package.
It is designed for reproducibility checks before publication.

```python
import weatherisk
from weatherisk.pipeline import run_pipeline
from weatherisk.parameters import get_preset

print("weatherisk imported successfully")
```

```python
preset = get_preset("stripes")
result = run_pipeline(
    resolution=10,
    n_sim=10,
    df=preset.df,
    alpha=preset.alpha,
    seed=42,
    output_dir="output/notebook_quickstart",
)

print("Number of clusters:", len(set(result["clusters"])))
print("Output directory: output/notebook_quickstart")
```
